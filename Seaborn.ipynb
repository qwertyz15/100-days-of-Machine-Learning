# Distribution plots
    1.distplot
    2.joinplot
    3.pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.describe()
```

```python
df['sex'].value_counts()
```

# Correlation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Distplot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde = False,bins = 10)
```

```python
sns.distplot(df['total_bill'],kde = False,bins = 10)
```

## Joint Plot

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = 'hex')
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = 'reg')
```

## Pair Plot

```python
sns.pairplot(df)
```

# Categorical Plots
    1.Count Plot
    2.Bar Plot
    3.Box Plot
    4.Violin Plot

## Count Plot

```python
sns.countplot('sex', data = df)
```

```python
sns.countplot(y = 'sex', data = df)
```

## Bar Plot

```python
sns.barplot(x = 'sex', y = 'size', data = df)
```

```python
sns.catplot(x = 'sex', y = 'size', data = df, kind = 'bar')
```

## Box Plot

```python
sns.boxplot(x = 'tip', y = 'sex', data = df)
```

```python
sns.catplot(x = 'tip', y = 'sex', data = df, kind = 'box')
```

```python
sns.catplot(data = df, kind = 'box', orient = 'v')
```

```python
sns.catplot(x = 'tip', y = 'sex', data = df, kind = 'box', hue = 'day')
```

```python
sns.catplot(x = 'tip', y = 'sex', data = df, kind = 'box', hue = 'day', row = 'smoker')
```

```python
sns.catplot(x = 'tip', y = 'sex', data = df, kind = 'box', hue = 'day', col = 'smoker')
```

## Violin Plot

```python
sns.violinplot(x = 'tip', y = 'sex', data = df)
```

```python
sns.catplot(x = 'tip', y = 'sex', data = df, kind = 'violin')
```
